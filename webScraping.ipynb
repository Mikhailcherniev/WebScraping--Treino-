```python
# web

# módulos para controlar o navegador web
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.common.exceptions import TimeoutException

# dataframe e tempo
import pandas as pd
import time

# Caminho para o ChromeDriver
chrome_driver_path = "C:\\Program Files\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe"

# Configuração do WebDriver
servico = Service(chrome_driver_path)
controle = webdriver.ChromeOptions()
controle.add_argument("--disable-gpu")
controle.add_argument("--window-size=1920,1080")

# Inicialização do WebDriver
driver = webdriver.Chrome(service=servico, options=controle)

# URL alvo
url_base = "https://masander.github.io/AlimenticiaLTDA-financeiro/?authuser=0"
driver.get(url_base)
time.sleep(5)  # Aguarda o carregamento inicial

# Dicionário para armazenar os dados
dick_armazenamento = {
    "pessoas": [], 
    "data": [], 
    "tipo": [], 
    "setor": [], 
    "valor": [], 
    "fornecedor": []
}

# Espera até que a tabela esteja presente
try:
    WebDriverWait(driver, 10).until(
        ec.presence_of_element_located((By.ID, "Table"))
    )
    print("Tabela carregada com sucesso!")
except TimeoutException:
    print("Tempo de espera excedido!")
    driver.quit()
    exit()

# Localiza a tabela e suas linhas
tabela = driver.find_element(By.ID, "Table")
linhas = tabela.find_elements(By.TAG_NAME, "tr")

# Percorre cada linha da tabela e extrai os dados
for linha in linhas:
    try:
    
        pessoas = linha.find_element(By.CLASS_NAME, "td_id_despesa").text.strip()
        data = linha.find_element(By.CLASS_NAME, "td_data").text.strip()
        tipo = linha.find_element(By.CLASS_NAME, "td_tipo").text.strip()
        setor = linha.find_element(By.CLASS_NAME, "td_setor").text.strip()
        valor = linha.find_element(By.CLASS_NAME, "td_valor").text.strip()
        fornecedor = linha.find_element(By.CLASS_NAME, "td_fornecedor").text.strip()

        print(f"{pessoas} - {setor}")

        dick_armazenamento["pessoas"].append(pessoas)
        dick_armazenamento["data"].append(data)
        dick_armazenamento["tipo"].append(tipo)
        dick_armazenamento["setor"].append(setor)
        dick_armazenamento["valor"].append(valor)
        dick_armazenamento["fornecedor"].append(fornecedor)
        
    except Exception as e:
        print("Erro ao coletar dados:", e)

# Fecha o navegador
driver.quit()

# Salva os dados em um arquivo Excel
df = pd.DataFrame(dick_armazenamento)
df.to_excel("dados_despesas.xlsx", index=False)

print(f"Arquivo 'dados_despesas.xlsx' salvo com sucesso! ({len(df)} registros capturados)")
```
