```python
# web

# módulos para controlar o navegador web
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.common.exceptions import TimeoutException

# dataframe e tempo
import pandas as pd
import time

# Caminho para o ChromeDriver
chrome_driver_path = "C:\\Program Files\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe"

# Configuração do WebDriver
servico = Service(chrome_driver_path)
controle = webdriver.ChromeOptions()
controle.add_argument("--disable-gpu")
controle.add_argument("--window-size=1920,1080")

# Inicialização do WebDriver
driver = webdriver.Chrome(service=servico, options=controle)

# URL alvo
url_base = "https://masander.github.io/AlimenticiaLTDA-financeiro/?authuser=0"
driver.get(url_base)
time.sleep(3)  # Aguarda o carregamento inicial

# CLICA NO BOTÃO "ORÇAMENTOS"
try:
    botao_orcamentos = WebDriverWait(driver, 10).until(
        ec.element_to_be_clickable((By.XPATH, "//button[contains(text(),'Orçamentos')]"))
    )
    botao_orcamentos.click()
    print("Botão 'Orçamentos' clicado com sucesso.")
    time.sleep(3)  # Aguarda a tabela aparecer após o clique
except TimeoutException:
    print("Botão 'Orçamentos' não foi encontrado ou não pôde ser clicado.")
    driver.quit()
    exit()

# Dicionário para armazenar os dados
dick_armazenamento = {
    "setor": [], 
    "mes": [], 
    "ano": [], 
    "valor_previsto": [], 
    "valor_realizado": [], 
}

# Espera até que a tabela esteja presente
try:
    WebDriverWait(driver, 10).until(
        ec.presence_of_element_located((By.ID, "Table"))
    )
    print("Tabela carregada com sucesso!")
except TimeoutException:
    print("Tempo de espera excedido ao tentar localizar a tabela!")
    driver.quit()
    exit()

# Localiza a tabela e suas linhas
tabela = driver.find_element(By.ID, "Table")
linhas = tabela.find_elements(By.TAG_NAME, "tr")

# Percorre cada linha da tabela e extrai os dados
for linha in linhas:
    try:
        setor = linha.find_element(By.CLASS_NAME, "td_setor").text.strip()
        mes = linha.find_element(By.CLASS_NAME, "td_mes").text.strip()
        ano = linha.find_element(By.CLASS_NAME, "td_ano").text.strip()
        valor_previsto = linha.find_element(By.CLASS_NAME, "td_valor_previsto").text.strip()
        valor_realizado = linha.find_element(By.CLASS_NAME, "td_valor_realizado").text.strip()
    

        print(f"{setor} - {ano}")

        dick_armazenamento["setor"].append(setor)
        dick_armazenamento["mes"].append(mes)
        dick_armazenamento["ano"].append(ano)
        dick_armazenamento["valor_previsto"].append(valor_previsto)
        dick_armazenamento["valor_realizado"].append(valor_realizado)
        
        
    except Exception as e:
        print("Erro ao coletar dados:", e)

# Fecha o navegador
driver.quit()

# Salva os dados em um arquivo Excel
df = pd.DataFrame(dick_armazenamento)
df.to_excel("dados_orcameto.xlsx", index=False)

print(f"Arquivo 'dados_orcamento.xlsx' salvo com sucesso! ({len(df)} registros capturados)")
```
